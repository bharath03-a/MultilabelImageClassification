```python
import pandas as pd
import matplotlib.pyplot as plt

resnet_metrics = pd.read_csv('./output_metrics/resnet_metrics.csv')
densenet_metrics = pd.read_csv('./output_metrics/denset_metrics.csv')

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 15))

metrics = ['precision', 'recall', 'f1_score']
for i, metric in enumerate(metrics):
    ax = axes[i]
    
    resnet_sorted = resnet_metrics.sort_values(by=metric, ascending=False)
    densenet_sorted = densenet_metrics.sort_values(by=metric, ascending=False)
    
    ax.bar(resnet_sorted['class_name'], resnet_sorted[metric], label='ResNet50', alpha=0.7)
    ax.bar(densenet_sorted['class_name'], densenet_sorted[metric], label='DenseNet121', alpha=0.7)
    
    ax.set_ylabel(metric.capitalize())
    ax.set_title(f'Testing Results - {metric.capitalize()} Comparison')
    plt.setp(ax.get_xticklabels(), rotation=90, ha='right')
    ax.legend()

plt.tight_layout()
plt.show()
```
